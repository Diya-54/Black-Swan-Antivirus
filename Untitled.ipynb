```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df_raw = pd.read_csv("features_data.csv")
df_raw.head()
```

```python
LABELED_CSV = 'features_labeled.csv'
LABELED_CSV
```

```python
import re

MALICIOUS_PATTERNS = [
    'meterpreter', 'shell_reverse', 'shell_bind', 'patchupmeterpreter',
    'vncinject', 'custom_reverse', 'custom_bind', 'pingback',
    'metsvc', 'adduser', 'format_all_drives', 'download_exec',
    'speak_pwned', 'messagebox', 'powershell_reverse', 'powershell_bind',
    'find_tag', 'reverse_tcp', 'bind_tcp', 'reverse_http', 'reverse_https',
    'reverse_udp', 'bind_udp', 'named_pipe', 'reverse_ord', 'bind_ord'
]
BENIGN_EXTENSIONS    = ['.txt', '.log', '.md', '.csv', '.json', '.so', '.o']
MALICIOUS_EXTENSIONS = ['.exe', '.dll', '.ps1', '.bat', '.vbs',
                         '.sys', '.scr', '.pif', '.com', '.elf', '.sh']
BENIGN_BINARY_NAMES  = [
    'ls', 'cat', 'grep', 'find', 'curl', 'wget', 'tar', 'zip',
    'gcc', 'python3', 'bash', 'sh', 'cp', 'mv', 'rm', 'chmod'
]

def auto_label(fname):
    fname_lower = fname.lower()

    # 1. Malicious by strong payload keywords ONLY
    for pat in MALICIOUS_PATTERNS:
        if pat in fname_lower:
            return 1

    # 2. Hash-named + suspicious extension → malicious
    name_only = fname_lower.rsplit('.', 1)[0]
    is_hash   = bool(re.match(r'^[0-9a-f]{32,}$', name_only))
    if is_hash:
        for ext in MALICIOUS_EXTENSIONS:
            if fname_lower.endswith(ext):
                return 1

    # 3. Known system binaries → benign
    if fname_lower in BENIGN_BINARY_NAMES:
        return 0

    # ❗ DO NOT auto-label benign by extension anymore
    return -1

# Apply
df = df_raw.copy()
df['label'] = df['filename'].apply(auto_label)
df['label'] = df['label'].replace(-1, 0)

print('Label distribution:')
print(df['label'].value_counts().rename({0:'Benign', 1:'Malicious', -1:'Unknown'}))


df.to_csv(LABELED_CSV, index=False)

print(f"Labeled CSV saved as: {LABELED_CSV}")
```

```python
df = df_raw.copy()
df['label'] = df['filename'].apply(auto_label)
```

```python
# Check what the 43 unknown filenames actually look like
unknown = df[df['label'] == -1][['filename']]
print(f"Total unknowns: {len(unknown)}")
print(unknown.to_string())
```

```python
df_clean = df.copy()

# Convert unknown → benign
df_clean['label'] = df_clean['label'].replace(-1, 0)

print(f'Training samples: {len(df_clean)} (including unknowns as benign)')
```

```python
DROP_COLS = ['filename', 'label']
X = df_clean.drop(columns=DROP_COLS)
y = df_clean['label']
```

```python
counts = y.value_counts()
ratio  = counts.min() / counts.max()
print(f'\nClass balance — Benign: {counts[0]}  |  Malicious: {counts[1]}')
if ratio < 0.3:
    print('⚠ Class imbalance detected — models will use class_weight="balanced"')
print(f'Feature matrix: {X.shape[0]} rows × {X.shape[1]} features')
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.25,
    random_state=42,
    stratify=y          # keeps class ratio the same in train & test
)
print(f'Train: {len(X_train)}  |  Test: {len(X_test)}')
```

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, ExtraTreesClassifier, VotingClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.metrics import (
    classification_report, accuracy_score, roc_auc_score,
    confusion_matrix, ConfusionMatrixDisplay
)
from xgboost import XGBClassifier
from sklearn.preprocessing import LabelEncoder
                  

import joblib
```

```python
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
scale_pos_weight = float(np.sum(y_train == 0)) / max(float(np.sum(y_train == 1)), 1)

models = {
    'Random Forest': RandomForestClassifier(
        n_estimators=200, class_weight='balanced', random_state=42, n_jobs=-1),
    'Extra Trees': ExtraTreesClassifier(
        n_estimators=200, class_weight='balanced', random_state=42, n_jobs=-1),
    'Gradient Boosting': GradientBoostingClassifier(
        n_estimators=150, learning_rate=0.1, max_depth=4, random_state=42),
    'XGBoost': XGBClassifier(
        n_estimators=300, learning_rate=0.05, max_depth=6,
        scale_pos_weight=scale_pos_weight, eval_metric='logloss',
        random_state=42, n_jobs=-1),
}
```

```python
results = {}
print(f'{"Model":<25} {"CV Acc (mean)":>15} {"CV Acc (std)":>14} {"CV F1":>10}')
print('-' * 68)
for name, model in models.items():
    cv_acc = cross_val_score(model, X_train, y_train, cv=cv, scoring='accuracy', n_jobs=-1)
    cv_f1  = cross_val_score(model, X_train, y_train, cv=cv, scoring='f1', n_jobs=-1)
    results[name] = {'cv_acc_mean': cv_acc.mean(), 'cv_acc_std': cv_acc.std(), 'cv_f1_mean': cv_f1.mean()}
    print(f'{name:<25} {cv_acc.mean():>14.4f} {cv_acc.std():>14.4f} {cv_f1.mean():>10.4f}')

print('\n✓ Cross-validation complete')
```

```python
from sklearn.utils import shuffle

y_random = shuffle(y_train, random_state=42)
random_scores = cross_val_score(
    models['Extra Trees'], X_train, y_random, cv=cv, scoring='f1', n_jobs=-1
)
print(f"Random label F1: {random_scores.mean():.4f}")
```

```python
best_name = 'Random Forest'  #i will use rf as its already written in literature survey.
best_model=models[best_name]
best_model.fit(X_train, y_train)
y_pred = best_model.predict(X_test)
```

```python
print(f'\nTest Accuracy : {accuracy_score(y_test, y_pred):.4f}')
try:
    prob = best_model.predict_proba(X_test)[:, 1]
    print(f'Test ROC-AUC  : {roc_auc_score(y_test, prob):.4f}')
except Exception:
    pass

print('\nClassification Report:')
print(classification_report(y_test, y_pred, target_names=['Benign', 'Malicious']))
```
